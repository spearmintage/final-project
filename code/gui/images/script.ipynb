```python
import requests
import pandas as pd
import random
import os
from PIL import Image

api_key = "3d863b47-d690-4ce8-a070-28c595e98c16"
```

```python
def run(start, end):
    new_data = []

    url = "https://nuthatch.lastelm.software/v2/birds"
    headers = {"api-key": api_key}

    for row in open("../species_key.csv").readlines()[start:end]:
        name = row.split(",")[1].replace("\n", "")

        params = {
            "name": name,
            "hasImg": "true"
        }

        json = requests.get(url, params, headers=headers).json()["entities"]

        new_data.append([name, json])

    return new_data
```

```python
# RUN AT 10AM

data = []

start, end = 0, 350
data += run(start, end)

data
```

```python
# RUN AT 11:02AM

start, end = 350, 700
data += run(start, end)

data
```

```python
# RUN AT 12:04PM

start, end = 700, 992
data += run(start, end)

data
```

```python
for species_combo in data.values:
    name = species_combo[0]
    json = species_combo[1]

    if not (os.path.exists(name + ".jfif") and os.path.exists(name + ".png")):
        if json != []:
            img_url = random.choice(json[0]["images"])

            img_data = requests.get(img_url).content
            with open(name + ".jfif", "wb") as image_writer:
                image_writer.write(img_data)

            jfif_img = Image.open(name + ".jfif")
            jfif_img.save(name + ".png")
```

```python
for file in os.listdir():
    if file[-4:] == "jfif":
        os.remove(file)
```

```python
pd.DataFrame(data).to_csv("species_image_data.csv")
```
