```python
# expected total files: 112861
# expected unique file names: 112861

wav_directory = "/mnt/lustre/peprmint/train_audio_wav/"
```

```python
import os
from tqdm import tqdm

total_files = 0
total_unique_files = 0

for subfolder in tqdm(os.listdir(wav_directory)):
    subfolder_files = os.listdir(wav_directory + subfolder)

    total_files += len(subfolder_files)
    total_unique_files += len(set(subfolder_files))

print(total_files, "total files.")
print(total_unique_files, "total unique files.")
```
