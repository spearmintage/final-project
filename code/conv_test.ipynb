```python
import torch
import torch.nn as nn
import numpy as np
```

```python
input = torch.ones(size=(4, 3, 6, 10))
input # 4x3x6x10 input
# output should be of shape 1xDx6x10 (D = output features)?
```

```python
# given input of shape AxBxCxD
# within conv2d: in_channels = B
# C and D are only reduced relative to kernel_size if padding is not mentioned
# A should also be kept the same

# given input of shape AxBxCxD
# where: 
#   out_channels = P
#   kernel_size = (Q, R)
#   padding = 0
# output will be of shape AxPx(C-Q+1)x(D-R+1)

conv1 = nn.Conv2d(in_channels=input.shape[1], out_channels=5, kernel_size=(2, 6))
output1 = conv1(input)
print("Conv 1:", conv1)
print("After Conv 1:")
print("Input:", input.shape)
print("Output:", output1.shape)
print(output1)
print()

conv2 = nn.Conv2d(in_channels=output1.shape[1], out_channels=3, kernel_size=3, stride=(2,1))
output2 = conv2(output1)
print("Conv 2:", conv2)
print("After Conv 2:")
print(output1.shape)
print(output2.shape)
print(output2)
```
