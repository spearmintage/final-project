this notebook converts all the files in /mnt/lustre/peprmint/train_audio from mp3s to wavs, in a separate folder: train_audio_wav

```python
import subprocess

subprocess.call(["ffmpeg", "-i", "/mnt/lustre/peprmint/train_audio/bkpwar/XC51862.mp3", "/mnt/lustre/peprmint/train_audio_wav/bkpwar/XC51862.wav"])
```

```python
import os
import soundfile as sf

from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
input_folder = 'ogg_files_folder'
output_folder = 'wav_files_folder'
```

```python
home_dir = "/mnt/lustre/peprmint/"

old_folder = home_dir + "train_audio/"
new_folder = home_dir + "train_audio_wav/"

if not os.path.exists(new_folder):
    os.makedirs(new_folder)

total_folders = len(os.listdir(old_folder))

erroneous_files = []

for i, subfolder in enumerate(os.listdir(old_folder)):
    if not os.path.exists(new_folder + subfolder):
        os.makedirs(new_folder + subfolder)

    if len(os.listdir(old_folder + subfolder)) == len(os.listdir(new_folder + subfolder)) and len(os.listdir(old_folder + subfolder)) != 0:
        # if old folder's subfolder contains the same files as new folder's subfolder and is NOT empty, ignore
        print(f"Skipping {subfolder}, as it is already full.") 
        continue
    for filename in (progress_bar := tqdm(os.listdir(old_folder + subfolder))):
        if filename.endswith('.ogg') or filename.endswith('.mp3'):
            new_file_path = f"{new_folder + subfolder}/{filename.split('.')[0]}.wav"
            if not os.path.isfile(new_file_path):
                # only make a new copy if the new file doesn't exist
                # no need to even read or overwrite the new file
                old_file_path = f"{old_folder}{subfolder}/{filename}"
                try:
                    data, samplerate = sf.read(old_file_path)
                    
                    sf.write(new_file_path, data, samplerate)
                except:
                    erroneous_files.append(old_file_path)
                    
                    print("Some error occurred while trying to read and write", subfolder + "/" + filename)
        progress_bar.set_description(f"Folder {subfolder} ({i + 1}/{total_folders}):")
```
