```python
import pandas as pd

folders = pd.DataFrame()
```

```python
import os

unique_files = []
files = []
folders = []

dataset_dir = "birdsongs-combined/train_audio"
for folder in os.listdir(dataset_dir):
    folder_path = dataset_dir + "/" + folder

    file_count = len(os.listdir(folder_path))
    unique_file_count = len(set(os.listdir(folder_path)))

    folders.append(folder)
    files.append(file_count)
    unique_files.append(unique_file_count)

folders = pd.Series(folders)
files = pd.Series(files)
unique_files = pd.Series(unique_files)

folder_df = pd.DataFrame((folders, files, unique_files), index=["folder", "files", "unique_files"]).transpose()
folder_df
```
