```python
import os
import pandas as pd
```

```python
train_audio_path = "train_audio/"

folder_sheet = pd.read_excel("metadata/final_metadata.xlsx", "FOLDER_SPECIES_TABLE")[["UNIQUE TEXT", "Unnamed: 7"]].iloc[1:1015, :]
folder_sheet.columns = ["folder_code", "species"]
folder_sheet
```

```python
try:
    folders = os.listdir(train_audio_path)
except:
    folders = []
folders
```

```python
# determines which species of the 1014 are not present in the train_audio directory.

for folder in folder_sheet["folder_code"]:
    if folder not in folders:
        print(folder, "not in train_audio folder.")
```
