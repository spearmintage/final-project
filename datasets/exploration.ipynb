```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

folders = pd.DataFrame()
```

```python
import os

unique_files = []
files = []
folders = []

dataset_dir = "birdsongs-combined/train_audio"
for folder in os.listdir(dataset_dir):
    folder_path = dataset_dir + "/" + folder

    file_count = len(os.listdir(folder_path))
    unique_file_count = len(set(os.listdir(folder_path)))

    folders.append(folder)
    files.append(file_count)
    unique_files.append(unique_file_count)

folders = pd.Series(folders)
files = pd.Series(files)
unique_files = pd.Series(unique_files)

folder_df = pd.DataFrame((folders, files, unique_files), index=["folder", "files", "unique_files"]).transpose()

folder_df["unique_diff"] = folder_df["files"] - folder_df["unique_files"]
unique_sum = folder_df["unique_files"].sum()
folder_df["unique_percentage"] = folder_df["unique_files"] * 100.0 / unique_sum

folder_df
```

```python
folder_df.sort_values(by="unique_files", ascending=False)
```

```python
# folders with less than 10 files

insignificant_folders = folder_df[folder_df["unique_files"] < 10].sort_values(by="unique_files")

print(insignificant_folders)

print(f"\nTotal percentage of files in folders with <10 files: {insignificant_folders["unique_percentage"].sum():.03f}%")
print("NOTE: this does NOT take into account file length. One file in a folder could be 5 hours long, while a folder with 50 files could all be 5 second clips.")
```

```python
# insights and analysis
plt.figure(figsize=(16, 9))
folder_df.sort_values(by="unique_files")["unique_files"].hist(bins=np.arange(0, max(folder_df["unique_files"]), 10))
plt.title("Distribution of Audio file Count Per Folder")
plt.xticks(np.arange(0, max(folder_df["unique_files"]), 50))
plt.xlabel("Files")
plt.ylabel("Count")
```
